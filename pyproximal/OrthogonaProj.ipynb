# Orthogonal projections

```python
%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt

from pyproximal.projection import *
```

## Box

```python
pop = BoxProj(-1, 1)

x = np.arange(-5, 5, 0.1)
xp = pop(x)

plt.plot(x, x, 'k')
plt.plot(x, xp, 'r');
```

```python
pop = BoxProj(-5, 1)

x = np.arange(-10, 10, 0.1)
xp = pop(x)

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```

```python
pop = BoxProj(upper=1)

x = np.arange(-10, 10, 0.1)
xp = pop(x)

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```

```python
pop = BoxProj(lower=1)

x = np.arange(-10, 10, 0.1)
xp = pop(x)

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```

```python
t = np.arange(-10, 10, 0.1)
x = t ** 2
l = 5*np.abs(np.arange(-10, 10, 0.1))
u = 8*np.abs(np.arange(-10, 10, 0.1))

pop = BoxProj(lower=l, upper=u)
xp = pop(x)

plt.plot(t, x, 'k');
plt.plot(t, l, '--k');
plt.plot(t, u, '--k');
plt.plot(t, xp, 'r');
```

## Hyperplane-box

```python
coeffs = np.array([1, 1, -1, -1])
hop = HyperPlaneBoxProj(coeffs, 0, -10000, 10000, maxiter=100, xtol=1e-10)

x = np.array([-1, 2, 3, 1])
xp = hop(x)
print(xp)

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```

```python
# using matrices
coeffs = np.array([[1, 1], [-1, -1]])
hop = HyperPlaneBoxProj(coeffs, 0, -np.inf, np.inf, maxiter=100, xtol=1e-10)

x = np.array([[-1, 2], [3, 1]])
xp = hop(x)
print(xp)

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```

## Simplex

```python
hop = SimplexProj(3, 1, maxiter=100, xtol=1e-10)

x = np.array([0, 0.5, -0.1])
xp = hop(x)
print(xp, np.sum(xp))

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```

```python
hop = SimplexProj(3, 2, maxiter=100, xtol=1e-10)

x = np.array([0, 0.5, -0.1])
xp = hop(x)
print(xp, np.sum(xp))

plt.plot(x, x, 'k');
plt.plot(x, xp, 'r');
```
